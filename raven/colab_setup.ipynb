```python
# Cell 4: Run the pipeline
import json
from raven.app.main import run_full_pipeline

# Option A: YouTube URL se
VIDEO_URL = "https://www.youtube.com/watch?v=dQw4w9WgXcQ"  # Change this URL

result = run_full_pipeline({"youtube_url": VIDEO_URL})
print(json.dumps(result, indent=2, ensure_ascii=False))
```

```python
# Cell 5: Download output clips
from google.colab import files

if result.get("status") == "success":
    for clip in result["data"]["final_outputs"]:
        path = clip.get("deliverable_path") or clip.get("video_path")
        if path and os.path.exists(path):
            files.download(path)
            print(f"✅ Downloaded: {path}")
else:
    print("❌ Pipeline failed. Check the error above.")
```
