```python
# Cell 4: RUN PIPELINE - Paste YouTube URL and generate shorts!
import json

VIDEO_URL = "https://www.youtube.com/watch?v=uX4vCuQO__0"  # <-- CHANGE THIS URL

print(f"\ud83d\ude80 Processing: {VIDEO_URL}")
print("This will take several minutes...")

result = run_full_pipeline({"youtube_url": VIDEO_URL})

if result.get("status") == "success":
    clips = result["data"]["final_outputs"]
    print(f"\n\ud83c\udf89 {len(clips)} shorts generated!")
    for c in clips:
        title = c.get("metadata", {}).get("title", "clip")
        print(f"  \ud83d\udcf9 {title}")
else:
    print("\u274c Failed:", result.get("error"))

print(json.dumps(result, indent=2, ensure_ascii=False))
```

```python
# Cell 5: Download clips to your computer
from google.colab import files
if result.get("status") == "success":
    for clip in result["data"]["final_outputs"]:
        path = clip.get("deliverable_path") or clip.get("video_path", "")
        if path and os.path.exists(path):
            files.download(path)
    print("\ud83d\udfe2 Download started! Check your browser downloads.")
else:
    print("Nothing to download - pipeline didn't succeed.")
```
