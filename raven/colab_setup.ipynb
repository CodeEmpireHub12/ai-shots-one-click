```python
# ============================================================
# 🚀 PROJECT RAVEN - ONE CLICK SHORTS GENERATOR
# Run this ONE cell. It will:
# 1. Install dependencies  2. Clone repo  3. Setup path
# 4. Ask for URL  5. Run pipeline  6. Auto-download shorts
# ============================================================

import os, sys, json, time, shutil
from IPython.display import clear_output, display, HTML

print("=" * 60)
print("🚀 PROJECT RAVEN - ONE CLICK SHORTS GENERATOR")
print("=" * 60)

# ─── Step 1: Install dependencies ───
print("\n[1/5] 📦 Installing dependencies...")
!pip install -q pyyaml opencv-python moviepy numpy scipy librosa soundfile faster-whisper scenedetect yt-dlp
print("       ✅ Done")

# ─── Step 2: Clone repo ───
print("[2/5] 📥 Cloning repository...")
if os.path.exists("/content/ai-shots-one-click"):
    shutil.rmtree("/content/ai-shots-one-click")
!git clone --quiet https://github.com/CodeEmpireHub12/ai-shots-one-click.git /content/ai-shots-one-click
print("       ✅ Done")

# ─── Step 3: Set Python path ───
print("[3/5] 🔧 Setting up Python path...")
sys.path.insert(0, "/content/ai-shots-one-click")
os.chdir("/content/ai-shots-one-click")
import raven
import raven.app
from raven.app.main import run_full_pipeline
print("       ✅ Done")

# ─── Step 4: Get YouTube URL ───
print("[4/5] 🎬 Ready!")
print("\n" + "-" * 60)
video_url = input("PASTE YOUTUBE URL HERE and press Enter: ").strip()

if not video_url:
    video_url = "https://www.youtube.com/watch?v=dQw4w9WgXcQ"
    print(f'⚠️ No URL entered. Using default video.')

print(f"\n🚀 Processing: {video_url}")
print("⏳ This takes 3-10 minutes depending on video length...")
print()

# ─── Step 5: Run pipeline ───
start_time = time.time()
try:
    result = run_full_pipeline({"youtube_url": video_url})
    elapsed = time.time() - start_time

    clear_output(wait=True)
    print("=" * 60)
    print("✅✅✅ PIPELINE COMPLETED SUCCESSFULLY! ✅✅✅")
    print(f"⏱️  Time taken: {elapsed:.0f}s ({elapsed/60:.1f} min)")
    print("=" * 60)

    if result.get("status") == "success":
        clips = result["data"]["final_outputs"]
        print(f"\n🎉 {len(clips)} SHORTS GENERATED!")

        for i, clip in enumerate(clips, 1):
            path = clip.get("deliverable_path") or clip.get("video_path", "")
            meta = clip.get("metadata", {})
            title = meta.get("title", f"Clip {i}")
            size = f"{os.path.getsize(path) / 1024 / 1024:.1f} MB" if path and os.path.exists(path) else "N/A"
            print(f"  📹 Short #{i}: {title} ({size})")

        # ─── Auto Download ───
        print(f"\n[5/5] 📥 Downloading {len(clips)} short(s) to your computer...")
        from google.colab import files
        for clip in clips:
            path = clip.get("deliverable_path") or clip.get("video_path", "")
            if path and os.path.exists(path):
                files.download(path)
                time.sleep(1)

        print("\n" + "=" * 60)
        print("🎬 ALL SHORTS DOWNLOADED! CHECK YOUR BROWSER DOWNLOADS FOLDER.")
        print("=" * 60)
        print("\n🔄 Want to make more? Just run this cell again with a new URL!")
    else:
        print(f"\n❌ Pipeline failed: {result.get('error', 'Unknown error')}")

except Exception as e:
    print(f"\n❌ Error: {e}")
    import traceback
    traceback.print_exc()
```
